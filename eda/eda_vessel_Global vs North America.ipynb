```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
# function to show map
def show_map(df):  
    df = df[df['is_fishing'] >= 0]
    fig = go.Figure(go.Densitymapbox(lat=df.lat, lon=df.lon, z=df.speed,radius=2))
    fig.update_layout(mapbox_style="carto-darkmatter", mapbox_center_lon=180)
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()
```

```python
# show combined vessel map
fishing_vessel_all_df = pd.read_csv('../data/combine_gear_v1.csv')
fishing_vessel_all_df.head(5)
show_map(fishing_vessel_all_df)
```
